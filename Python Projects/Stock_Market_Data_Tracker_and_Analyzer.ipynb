# **Stock Market Data Tracker and Analyzer**

**Project Overview**

This project involved conducting in-depth research on multiple companies to analyze their stock performance.
```python
import pandas as pd #retrieves the necessary package requires to conduct the project
import yfinance as yf #imports the yfinace libaray
import matplotlib as plt
import datetime as dt
import matplotlib.pyplot as plt
import seaborn as sns  # Imports the seaborn library to create plots
```

#**Nike Stock Analysis**

*Short Term Analysis*

```python
ticker_symbol = "NKE" #defines the ticker symbol of the company we want to analyze (NIKE)
stock_data = yf.Ticker(ticker_symbol) #gets the stock data for NIKE
hist_data_20 = stock_data.history(period="20d")# defines what period we want to get the data for; shorterm 20 days
display(hist_data_20.head()) #displays the stock data in a table with the headings
```

```python
hist_data_20['Daily Return'] = hist_data_20['Close'].pct_change() #Adds a daily return column to the table
display(hist_data_20)
```

```python
hist_data_20.fillna('-', inplace=True)# Replace NaN values with a hyphen ('-') using .fillna()
display(hist_data_20)# Display the new table after replacing NaN values
```

```python
%matplotlib inline
```

```python
#Creates a plot for the stock data over 20 days

plt.figure(figsize=(20,6)) #determines the size of the graph
plt.plot(hist_data_20['Close'],color='b',label='Nike stock over 20 days',linewidth=5)# sets the information to be plotted, the legend label and the clolor of the line
plt.title('Nike\'s Closing Price Over 20 Days')# Adds the title to the graph
plt.xlabel('Closing Date')# Labels the x axis
plt.ylabel(' Closing price ')# Labels the y axis
plt.legend()# Adds a legend to the plot
plt.xticks(rotation=45)# Rotates the x axis so we can see the infromation more clearly
plt.tight_layout() # Adjust layout to prevent overlapping of any labels
plt.show() #Displays the plot
```

```python
#Plots the volume traded in the last 20 days

volume_traded = hist_data_20['Volume']# gets what infromation needs to be plotted
plt.figure(figsize=(20,6))# determines the size of the plot
plt.plot(volume_traded, color='r',label='Volume traded over 20 days',linewidth=5)#plots the volume trades and sets the color labels the legend
plt.xlabel('Date')#Labels the x axis
plt.ylabel('Volume Traded')# Labels the y axis
plt.title('Nike\'s Volume Traded Over 20 Days')# Adds the title of the plot
plt.legend()# Adds a legend to the plot
plt.xticks(rotation=45)# Rotates the x axis so we can see the infromation more clearly
plt.tight_layout() # Adjust layout to prevent overlapping of any labels
plt.show() #Displays the plot
```

```python
# Plots the 30 and 300 days moving average for the last 20 days

hist_data_20['30_Move_Avg'] = hist_data_20['Close'].rolling(window=30).mean() # Calculates the 30 day moving average for the closing prices over the last 20 days
hist_data_20['300_Move_Avg'] = hist_data_20['Close'].rolling(window=300).mean() # Calculates the 300 day moving average for the closing prices over the last 20 days
plt.figure(figsize=(20, 6)) #determines the size of the plot
plt.plot(hist_data_20['Close'], label='Close Price',color='b',linewidth=5)#Plots the  closing prices from the last 20 days,sets the label for the legend and sets the color of the line
plt.plot(hist_data_20['30_Move_Avg'], label='30 day Moving Average',color='violet',linewidth=5)#Plots the  30 day moving average,sets the label for the legend and sets the color of the line
plt.plot(hist_data_20['300_Move_Avg'], label='300 Moving Average',color='yellow',linewidth=5)#Plots the  300 day moving average,sets the label for the legend and sets the color of the line
plt.xlabel('Date') # Labels the x-axis
plt.ylabel('Price') # Labels the y-axis
plt.title('Moving Averages of Nike\'s Stock over the last 20 days ') # Adds the plot title
plt.xticks(rotation=45)# Rotates the x axis so we can see the infromation more clearly
plt.legend()# Adds a legend
plt.show()# Displays the  plot
```

```python
#Finds the max price over the last 20 days
max_price_20 = hist_data_20['Close'].max()
print(f"The max closing price after 20 days  is ${max_price_20:.2f}")
```

```python
#Finds the min price over the last 20 days
min_price_20 = hist_data_20['Close'].min()
print(f"The min closing price after 20 days is ${min_price_20:.2f}")
```

```python
#Finds the percentage price over the last 20 days
percentage_change_20= (hist_data_20['Close'][-1]-hist_data_20['Close'][0]/hist_data_20['Close'][0]*100)
print(f"The percent change after 20 days is {percentage_change_20:.2f}%")
```

```python
#Finds the average price over the last 20 days
average_price_20 = hist_data_20['Close'].mean()
print(f"The average closing price after 20 days is ${average_price_20:.2f}")
```

```python
#Finds the median price over the last 20days
median_price_20 = hist_data_20['Close'].median()
print(f"The median closing price after 20 days  is ${median_price_20:.2f}")
```

# **Hyatt Stock Analysis**

*Short Term Analysis*

```python
ticker_symbol = "H" #defines the ticker symbol of the company we want to analyze (Hyatt)
stock_data = yf.Ticker(ticker_symbol) #gets the stock data for Hyatt
hist_data_H20 = stock_data.history(period="20d")# defines what period we want to get the data for; shorterm 30 days
display(hist_data_H20.head()) #displays the stock data in a table with the headings
```

```python
hist_data_H20['Daily Return'] = hist_data_H20['Close'].pct_change() #Adds a daily return column to the table
display(hist_data_H20.head())
```

```python
hist_data_H20.fillna('-', inplace=True)# Replace NaN values with a hyphen ('-') using .fillna()
display(hist_data_H20)# Display the new table after replacing NaN values
```

```python
#Creates a plot for the Hyatt stock data over 30 days

plt.figure(figsize=(20,6)) #determines the size of the graph
plt.plot(hist_data_H20['Close'],color='green',label='Hyatt stock over 20 days',linewidth=5)# sets the information to be plotted, the legend label and the clolor of the line
plt.title('Hyatt\'s Closing Price Over 20 Days')# Adds the title to the graph
plt.xlabel('Closing Date')# Labels the x axis
plt.ylabel(' Closing price')# Labels the y axis
plt.legend()# Adds a legend to the plot
plt.xticks(rotation=45)# Rotates the x axis so we can see the information more clearly
plt.tight_layout() # Adjust layout to prevent overlapping of any labels
plt.legend()
plt.show() #Displays the plot
```

```python
# Plots the volume traded in the last 20 days

volume_traded = hist_data_H20['Volume']  # gets the volume traded information
plt.figure(figsize=(20, 6))  # determines the size of the plot
sns.barplot(x=volume_traded.index, y=volume_traded, color='c',linewidth = 5)  # plots the volume trades as a bar plot using seaborn
plt.xlabel('Date')  # Labels the x axis
plt.ylabel('Volume Traded')  # Labels the y axis
plt.title('Hyatt\'s Volume Traded Over 20 Days')  # Adds the title of the plot
plt.xticks(rotation=45)  # Rotates the x axis so we can see the information more clearly
plt.tight_layout()  # Adjust layout to prevent overlapping of any labels
plt.show()  # Displays the plot
```

```python
# Plots the 30 and 300 days moving average for the last 20 days

hist_data_H20['30_Move_Avg'] = hist_data_H20['Close'].rolling(window=30).mean()# Calculates the 30-day moving average for the closing prices over the last 30 days
hist_data_H20['300_Move_Avg'] = hist_data_H20['Close'].rolling(window=300).mean()# Calculates the 300-day moving average for the closing prices over the last 30 days

plt.figure(figsize=(20, 6)) # determines the size of the plot
plt.plot(hist_data_H20['Close'], label='Close Price', color='green',linewidth = 5) # Plots the closing prices from the last 30 days, sets the label for the legend, and sets the color of the line
plt.plot(hist_data_H20['30_Move_Avg'], label='30 day Moving Average', color='c',linewidth = 5) # Plots the 30 day moving average, sets the label for the legend, and sets the color of the line
plt.plot(hist_data_H20['300_Move_Avg'], label='300 day Moving Average', color='brown',linewidth = 5) # Plots the 300 day moving average, sets the label for the legend, and sets the color of the line
plt.xlabel('Date') # Labels the x-axis
plt.ylabel('Price') # Labels the y-axis
plt.title('Moving Averages of Hyatt\' Stock over the last 20 days') # Adds the plot title
plt.legend() # Adds a legend
plt.show() # Displays the plot
```
