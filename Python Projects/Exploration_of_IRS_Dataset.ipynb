 **Project Overview:**

This mini project involves exploring the Iris dataset using Python libraries like Pandas, NumPy, Matplotlib, and Scikit-Learn. We load the data, manipulate it to add a new feature for petal area, and conduct statistical analysis to understand sepal lengths. Visualization via scatter plots helps us grasp relationships between sepal length and width. Finally, we train a logistic regression model to predict iris species based on sepal measurements and evaluate its accuracy. This project offers hands-on experience in data exploration, manipulation, visualization, and machine learning techniques.

```powershell
# Imports the required libaries needed

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix,accuracy_score
from sklearn.model_selection import train_test_split
```

```powershell
Iris_df = pd.read_csv('Iris.csv') # Loads the Iris dataset using pandas
Iris_df.head(5) # Displays the first five rows
```

```powershell
Iris_df.dtypes # Gets the datatypes of the columns
```

```powershell
Iris_df['Petal Area'] = Iris_df['PetalLengthCm'] * Iris_df['PetalWidthCm'] # Creates a new column 'Petal Width' by multiplying petal area by petal length
Iris_df.head(10) # Prints the first 10 rows
```

```powershell
Sepal_array = np.array(Iris_df['SepalLengthCm']) # Converts the Sepal Length column into an array
print(Sepal_array) # Prints the array
print()
Sepal_mean = Sepal_array.mean() # Calculates the mean
print(f"The speal mean is {Sepal_mean:.2f}") # Prints the mean accuarte to two decimal places
```

```powershell
Speal_std = np.std(Sepal_array) # Calculates the Sepal Standard Deviation
print(f"The speal std is {Speal_std:.2f}") # Prints the std accurate to two decimal paces
```

```powershell
Sepal_length = Iris_df['SepalLengthCm'] # Gets the sepal lengths
Sepal_Width = Iris_df['SepalWidthCm'] # Gets the sepal widths

plt.figure(figsize=(10,5)) # Size of plot
plt.scatter(Sepal_length, Sepal_Width, color='hotpink',marker = '*',label ='Sepal Length in relation to Sepal width') # Plots the legth and width in hot pink with stars as markers
plt.title('Sepal Length and Sepal Width') # Adds a title
plt.xlabel('Sepal Length (cm)') # Adds a x label
plt.ylabel('Sepal Width (cm)') # Adds a y label
plt.legend() # Adds a legend
plt.grid() # Adds a grid
plt.show() # Displays the plot
```

```powershell
X = Iris_df.iloc[:,[1,2]].values  #  sepal length and sepal width
y = Iris_df.iloc[:,[5]].values  # Species

#Prints the selected features and the label for the model
print('Selected features')
print(X[:5]) # displays the firdt 5
print()
print('Selected Lables ')
print(y[:5]) # Displays the first 5
print()

y = np.ravel(y) #.ravel puts the y as a one dimensional shape (a message came up after a runnoing recomming to do this)
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size = 0.2, random_state = 0)  #Splits the data into training and testing sets
classifier = LogisticRegression(random_state = 0) #Initializes the Logistic Regression Model
classifier.fit(X_train,y_train) #Trains the model
y_pred = classifier.predict(X_test) # Predicts the class labels for the test set
accuracy = accuracy_score(y_test,y_pred) # Calculates the accuracy scrore of the model

print(pd.DataFrame({'Actual':y_test,'Predicted':y_pred}).head(10)) # Prints the df actual vs preficted
print()
print(f"The accuracy of the model is: {accuracy:.2f}") # Prints the accuracy score of the model

```
